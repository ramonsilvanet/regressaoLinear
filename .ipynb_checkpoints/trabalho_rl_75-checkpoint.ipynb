```r
sugar_convert = data.frame(
    temp = c(1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.5, 1.7, 1.8, 1.9, 2.0),
    sugar = c(8.1, 7.8, 8.5, 9.8, 9.5, 8.9, 8.6, 10.2, 9.3, 9.2, 10.5)
)
```

```r
sugar_convert.lm = lm(sugar ~ temp, data = sugar_convert )
sugar_convert.coeffs = coefficients(sugar_convert.lm)
sugar_convert.coeffs
sugar_convert.res = resid(sugar_convert.lm)
```

```r
temp = 1.75
sugar_est = sugar_convert.coeffs[1] + sugar_convert.coeffs[2] * temp 
sugar_est
```

```r
plot(sugar_convert$temp, 
     sugar_convert.res,
     ylab="Residuals",
     xlab="temperature.",
     main="Sugar Convert Residuals"
    )
abline(0, 0)  
```
