```r
acid_extraction = data.frame(
    x = c(28, 48, 55, 75, 100, 109, 123, 138, 159, 164),
    y = c(43, 37, 62, 58, 66, 70, 76, 82, 88, 88)
)
```

```r
plot(acid_extraction$x, 
     acid_extraction$y,
     main = 'Acid Extraction',
     xlab = 'x',
     ylab = 'y')
```

```r
acid_extraction.lm = lm(y ~ x, data = acid_extraction )
acid_extraction.coeffs = coefficients(acid_extraction.lm)
acid_extraction.coeffs
acid_extraction.res = resid(acid_extraction.lm)
```

```r
plot(acid_extraction$x, 
     acid_extraction$y,
     main = 'Acid Extraction',
     xlab = 'x',
     ylab = 'y')
abline(acid_extraction.lm, col = 'blue')
```
