```r
chemical = data.frame(
    temp = c(0, 15, 30, 45, 60, 75),
    grams = c(7.33333333333333, 12, 23.3333333333333, 30.6666666666667, 41.6666666666667, 47.6666666666667)
)

#grams sum (20, 36, 70, 92, 125, 143)
#grams mean(6.6667, 12, 23.3334, 30.6667, 41.6667, 47.6667)
```

```r
chemical.lm = lm(grams ~ temp, data = chemical)
chemical.coeffs = coefficients(chemical.lm)
chemical.res = resid(chemical.lm)
```

```r
chemical.coeffs
```

```r
tempeature = 50
est_solve = chemical.coeffs[1] + chemical.coeffs[2] * tempeature 
est_solve
```

```r
plot(chemical$temp,
     chemical$grams, 
     main = 'Chemical Solve by Tempeature',
     ylab = 'Compound Dissolved (g)',
     xlab = 'Tempeature (ºC)')

abline(chemical.lm, col = 'blue')
```
