```r
strength_test_data = data.frame(
    arm_strength_x = c(17.3, 19.3, 19.5, 19.7, 22.9, 23.1, 26.4, 26.8, 27.6, 28.1, 28.2, 28.7, 29, 29.6, 29.9, 29.9, 30.3, 31.3, 36, 39.5, 40.4, 44.3, 44.6, 50.4, 55.9),
    dynamic_lift_y = c(71.7, 48.3, 88.3, 75, 91.7, 100, 73.3, 65, 75, 88.3, 68.3, 96.7, 76.7, 78.3, 60, 71.7, 85, 85, 88.3, 100, 100, 100, 91.7, 100, 71.7)
)
```

```r
strength_test_data.lm = lm(dynamic_lift_y ~ arm_strength_x, data = strength_test_data )
strength_test_data.coeffs = coefficients(strength_test_data.lm)
strength_test_data.coeffs
strength_test_data.res = resid(strength_test_data.lm)
```

```r
plot(strength_test_data$arm_strength_x, 
     strength_test_data$dynamic_lift_y,
     main = 'Arm Strength x Dynamic Lift',
    xlab = 'Strength Arm',
    ylab = 'Dynamic Lift')

abline(strength_test_data.lm, col = 'blue')
text(mean(strength_test_data$arm_strength_x) + 10, mean(strength_test_data$dynamic_lift_y), "y = 64.5292 + 0.5609x")
```

```r
dynamic_lift = 30
estimated_arm_strrength = strength_test_data.coeffs[1] + strength_test_data.coeffs[2] * dynamic_lift
estimated_arm_strrength
```
